The `plot_figures/Fig8_data` directory contains single-planet and two-planet Kepler-system map files. Files named `MaxEcc_16.txt` are single-planet slices; files named `MaxEcc_2pl_16.txt` are two-planet packing slices.

```python
from pathlib import Path
import matplotlib.pyplot as plt
from cbp_plotting import read_maxecc_file, plot_phase_map

single = read_maxecc_file('plot_figures/Fig8_data/MaxEcc_16.txt')
two_planet = read_maxecc_file('plot_figures/Fig8_data/MaxEcc_2pl_16.txt')
print(single.shape, two_planet.shape)
```

```python
# The single-planet Kepler files use a_0 and e_0 rather than mean anomaly on the second axis.
# The generic phase-map plotter can still show the grid structure, but publication-style Kepler maps
# may need custom labels/overlays for each system.
fig, ax = plot_phase_map(single, mode='delta_e', title='Kepler-16 single-planet slice')
ax.set_xlabel(r'$e_p$')
ax.set_ylabel(r'$a_p$ (AU)')
plt.show()
```

For publication-style overlays, start from `plot_figures/plot_Fig8_single.py` or the optional REBOUND slice scripts in `rebound_slices/`. The modernized notebooks avoid hard-coding every Kepler-system axis range so users can adapt the plot to new CBPs.
