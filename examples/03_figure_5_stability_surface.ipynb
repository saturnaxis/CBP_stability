This notebook plots the interpolated $a_c(\mu,e_{bin})$ surface and overlays the Kepler circumbinary planets used in the paper.

```python
import numpy as np
import matplotlib.pyplot as plt
from cbp_stability import build_stability_grid

grid = build_stability_grid()

names = ['16','34','35','38','47','64','413','453','1647']
mu = np.array([0.20255/(0.20255+0.68970), 1.0208/(1.0208+1.0479), 0.8094/(0.8094+0.8877), 0.249/(0.249+0.949), 0.342/(0.342+0.957), 0.408/(0.408+1.528), 0.5423/(0.5423+0.82), 0.1951/(0.1951+0.944), 0.9678/(0.9678+1.2207)])
e_b = np.array([0.15944,0.52087,0.1421,0.1032,0.0288,0.2117,0.0365,0.0524,0.1602])

fig, ax = plt.subplots(figsize=(9, 6), dpi=150)
mesh = ax.pcolormesh(grid.mu_values, grid.eb_values, grid.ac_values.T, shading='auto')
for name, x, y in zip(names, mu, e_b):
    ax.plot(x, y, 'wo', ms=5)
    ax.text(x, y+0.02, name, color='w', ha='center')
ax.set_xlabel(r'$\mu$')
ax.set_ylabel(r'$e_{bin}$')
ax.set_ylim(0, 0.8)
cbar = fig.colorbar(mesh, ax=ax)
cbar.set_label(r'$a_c/a_{bin}$')
plt.show()
```
