Figure 6 is a population-level dynamical-spacing summary rather than a raw archive-map figure. The table below uses the published values of the observed planet, critical limit, and spacing $eta$ to reproduce the main diagnostic: which systems have room for an equal-mass interior planet.

```python
import pandas as pd
import matplotlib.pyplot as plt

rows = [
    ('Kepler-16b', 0.6050, 0.7048, 2.4610),
    ('Kepler-34b', 0.8118, 1.0896, 7.1703),
    ('Kepler-35b', 0.4795, 0.6035, 6.3175),
    ('Kepler-38b', 0.4328, 0.4644, 1.1968),
    ('Kepler-47b', 0.1848, 0.2956, 13.519),
    ('Kepler-64b', 0.5368, 0.6340, 2.9697),
    ('Kepler-413b', 0.2389, 0.3530, 8.3487),
    ('Kepler-453b', 0.4184, 0.7903, 20.152),
    ('Kepler-1647b', 0.3497, 2.7200, 20.275),
]
df = pd.DataFrame(rows, columns=['system', 'a_c_au', 'a_p_au', 'beta'])
df['room_for_equal_mass_inner'] = df['beta'] >= 5
print(df)

fig, ax = plt.subplots(figsize=(9, 5), dpi=150)
df.plot.bar(x='system', y='beta', ax=ax, legend=False)
ax.axhspan(5, 7, alpha=0.15, label='Typical packing threshold')
ax.set_ylabel(r'Dynamical spacing $eta$')
ax.set_xlabel('')
ax.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
