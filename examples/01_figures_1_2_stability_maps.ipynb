The repository includes the 4-by-4 data subsets used for the Figure 1 and Figure 2 style maps. Each file has columns `a_0`, `M_0`, `emax`, `emin`, and `tcol`. Figure 1 corresponds to the binary beginning at periastron; Figure 2 uses the apastron comparison subset.

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from cbp_plotting import critical_ac_from_phase_grid, plot_phase_map, read_maxecc_file

path = Path('plot_figures/Fig1_data/MaxEcc_[0.3,0.3].txt')
data = read_maxecc_file(path)
print(data.shape)
print(f'a_c = {critical_ac_from_phase_grid(data):.3f}')
```

```python
fig, ax = plot_phase_map(data, mode='emax', title=r'Example: $\mu=0.3$, $e_{bin}=0.3$')
plt.show()
```
