The full simulation archive is not stored in GitHub because it is large. Download `MaxEcc.tar.gz` from Zenodo and place it in the repository root. The helper functions can read one member directly from the tarball without extracting the full archive.

```python
from pathlib import Path
from cbp_plotting import read_maxecc_from_tar, plot_phase_map, plot_archive_two_panel

archive = Path('MaxEcc.tar.gz')
if archive.exists():
    data = read_maxecc_from_tar(archive, 0.200, 0.200)
    fig, ax = plot_phase_map(data, mode='emax', title=r'$\mu=0.200$, $e_{bin}=0.200$')
else:
    print('Download MaxEcc.tar.gz from Zenodo and place it in the repository root before running this cell.')
```
